## formula para calcular area

## Datos de entrada

## Diagrama de flujo en lucichat
![Logo de Markdown](https://raw.githubusercontent.com/KarlaSaenz128/EPPL-Area-y-pertimetro-de-un-circulo-mis-imagenes/refs/heads/main/Diagrama%20ghihub.png)


```python
# Programa para calcular el área de un triángulo rectángulo
print("--- CALCULADORA DE ÁREA ---")
base = float(input("Ingresa el valor de la base (b): "))
altura = float(input("Ingresa el valor de la altura (h): "))
area = (base * altura) / 2
print(f"\nEl área del triángulo con base {base} y altura {altura} es: {area}")
```
